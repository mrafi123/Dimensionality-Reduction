```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
import os
os.environ['KERAS_BACKEND']='theano'
```

```python
from keras.datasets import mnist
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
print(X_train.shape)
```

```python
x_train = X_train.astype('float32') / 255.
x_test = X_test.astype('float32') / 255.
x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))

print(x_train.shape)
print(x_test.shape)
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
from sklearn.cluster import KMeans
est = KMeans(n_clusters=10,max_iter=500)
clusters=est.fit_predict(x_train)
est.cluster_centers_.shape
```

```python
fig=plt.figure(figsize=(28,13))
for i in range(10):
    ax=fig.add_subplot(5,7 ,1+i,xticks=[],yticks=[])
    ax.imshow(est.cluster_centers_[i].reshape((28,28)),cmap=plt.cm.binary)
    
plt.show()
```

```python
print(est.cluster_centers_[0].shape)
```

```python
print(clusters)
for i in range(60000):
    if clusters[i]== 0 :
        print(i)
```

```python
from scipy.stats import mode 
labels=np.zeros_like(clusters)
```

```python
print(y_train[0])
print(mode(y_train[0])[0])
```

```python
#print(labels)
for i in range(10):
    mask=(clusters==i)
    print(mask)
    labels[mask]= mode(y_train[mask])[0]
    print(labels[mask])
    #print(labels[i])
print(labels)
```

```python
from sklearn.decomposition import PCA
X=PCA(2).fit_transform(x_train[0:60000,])
kwargs=dict(cmap=plt.cm.get_cmap('rainbow',10),edgecolor='none',alpha=0.6)
fig,ax=plt.subplots(1,2,figsize=(8,4))
ax[0].scatter(X[:,0],X[:,1],c=labels,**kwargs)
ax[0].set_title('cluster labels')
ax[1].scatter(X[:,0],X[:,1],c=y_train,**kwargs)
ax[1].set_title('labels')
```

```python
print(x_train[0:60000,])
```

```python
plt.show()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_train,labels)
```

```python
from sklearn.cluster import KMeans
clusters_test=est.predict(x_test)
est.cluster_centers_.shape
```

```python
from scipy.stats import mode 
labels_test=np.zeros_like(clusters_test)
print(labels_test.shape)
```

```python
#print(y_test.shape)
for j in range(10):
    mask1=(clusters_test==j)
    print(mask1)
    labels_test[mask1]= mode(y_test[mask1])[0]
    print(labels_test[mask1])
    
print(labels_test)
```

```python
from sklearn.decomposition import PCA
X1=PCA(2).fit_transform(x_test[0:10000,])
kwargs=dict(cmap=plt.cm.get_cmap('rainbow',10),edgecolor='none',alpha=0.6)
fig,ax=plt.subplots(1,2,figsize=(8,4))
ax[0].scatter(X1[:,0],X1[:,1],c=labels_test,**kwargs)
ax[0].set_title('cluster_test labels')
ax[1].scatter(X1[:,0],X1[:,1],c=y_test,**kwargs)
ax[1].set_title('actual test labels')
plt.show()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,labels_test)
```

```python
import pandas as pd
df_train=pd.DataFrame({'assigned_labels':labels, 'actual_labels':y_train})
print(df_train.columns)
```
